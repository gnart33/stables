```python
import pandas as pd
```

```python
from pathlib import Path
root_path  = Path.cwd().parent
print(root_path)
# df = pd.read_csv(data_path / 'data.csv')
```

```python
df = pd.read_csv(root_path / 'data/processed/chain_data_2025-03-30.csv')
df.head()
```

# per chain

```python
df_chain = df.groupby('chain').agg({'current': 'sum', 'prevDay': 'sum'}).reset_index()
df_chain['change_day'] = df_chain['current'] - df_chain['prevDay']
df_chain['change_day_pct'] = (df_chain['change_day'] / df_chain['prevDay'])*100

df_chain.sort_values('change_day_pct', ascending=False)
```

```python
df_id = df.groupby('name').agg({'current': 'sum', 'prevDay': 'sum'}).reset_index()
df_id['change_day'] = df_id['current'] - df_id['prevDay']
df_id['change_day_pct'] = (df_id['change_day'] / df_id['prevDay'])*100
df_id.sort_values('change_day_pct', ascending=False)

df_id.sort_values('change_day_pct', ascending=False)

```
